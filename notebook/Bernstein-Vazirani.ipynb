```python
from qiskit import *
%matplotlib inline
from qiskit.tools.visualization import plot_histogram
```

```python
secret = '100101'
```

```python
l = len(secret)
circuit = QuantumCircuit(l+1, l)
circuit.h(range(l))
circuit.x(l)
circuit.h(l)
circuit.barrier()
circuit.draw(output='mpl')
```

```python
for ii, yesno in enumerate(reversed(secret)):
    if yesno == '1':
        circuit.cx(ii, l)
        
circuit.barrier()
circuit.draw(output='mpl')
```

```python
circuit.h(range(l))
circuit.barrier()

circuit.draw(output='mpl')
```

```python
circuit.measure(range(l), range(l))
```

```python
circuit.draw(output='mpl')
```

```python
simulator = Aer.get_backend('qasm_simulator')
res = execute(circuit, backend=simulator, shots=1).result()
counts = res.get_counts()
print(counts)
```
